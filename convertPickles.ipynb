```python
import pickle
import pandas as pd
import os
```

```python
def ConvertFiles(directory):
    for file in os.listdir(directory):
        filename = os.fsdecode(file)
        if filename.endswith(".pkl"): 
            full_path = os.path.join(directory, filename)
            print(full_path)
            with open(full_path, 'rb') as f:
                p = pickle.load(f)

            df = pd.DataFrame(p)
            print(full_path.removesuffix(".pkl") + '.csv')
            df.to_csv(full_path.removesuffix(".pkl") + '.csv')
            # print(df.head())
            continue
        else:
            continue
```
